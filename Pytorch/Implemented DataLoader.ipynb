```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['Unnamed: 32', 'id'], inplace=True)
df.head()
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df.iloc[:,0], test_size=0.2)
```

## Scaling

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
y_train
```

## Label Encoding

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

## Defining the Model

```python
class MyNN(nn.Module):

  # def  __init__(self, X):
  #   self.weights = torch.rand(X.shape[1], 1, dtype= torch.float64, requires_grad=True)
  #   self.bias = torch.zeros(1,dtype=torch.float64, requires_grad=True)

  def __init__(self,num_features):
    super().__init__()
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  # def forward(self, X):
  #   z = torch.matmul(X, self.weights) + self.bias
  #   y_pred = torch.sigmoid(z)
  #   return y_pred

  def forward(self, features):
    out = self.linear(features)
    y_pred = self.sigmoid(out)
    return y_pred

  # def loss_function(self, y_pred, y):
  #   # Clamp predictions to avoid log(0)
  #   eplsilon = 1e-7;
  #   y_pred = torch.clamp(y_pred, eplsilon, 1-eplsilon)

  #   # Calculate loss
  #   loss = - (y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred))
  #   return loss.mean()
```

### Important Parameters


```python
learning_rate = 0.1
epochs = 25

loss_function = nn.BCELoss()
```

## Data Loader & DataSet class


```python
from torch.utils.data import Dataset

class CustomDataset(Dataset):
  def __init__(self, data, labels, transform=None):
    self.data = data
    self.labels = labels
    self.transform = transform

  def __len__(self):
    return self.data.shape[0] # Return the total number of samples

  def __getitem__(self, index):
    sample = self.data[index]
    label = self.labels[index]
    if self.transform:
      sample = self.transform(sample) # Apply any transformations if provided
    return sample, label

from torch.utils.data import DataLoader

train_dataset = CustomDataset(X_train_tensor,  y_train_tensor)
test_dataset = CustomDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

### Training Pipeline


```python
# create model
model = MyNN(X_train_tensor.shape[1])

optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)

for epoch in range(epochs):

  for features, labels in train_loader:

    # Forward Bias
    # y_pred = model.forward(X_train_tensor)
    y_pred = model(features)

    # Loss
    loss = loss_function(y_pred, labels.view(-1,1))

    optimizer.zero_grad()

    # Backward pass
    loss.backward()

    optimizer.step()

    # parameters update
    # with torch.no_grad():
    #   model.linear.weight -= learning_rate * model.linear.weight.grad
    #   model.linear.bias -= learning_rate * model.linear.bias.grad

    # zero gradients
    # model.linear.weight.grad.zero_()
    # model.linear.bias.grad.zero_()

  print(loss.dtype)
  # print loss in each epoch
  print(f'Epoch: {epoch+1}, Loss: {loss.item()}')
```

## Evaluation

```python
# model evaluation
with torch.no_grad():
  y_pred = model.forward(X_test_tensor.float()) # This will give the values of y_pred from 0 to 1
  y_pred = (y_pred >0.9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy}')
```
